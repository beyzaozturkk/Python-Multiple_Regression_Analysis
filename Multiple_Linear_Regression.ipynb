# **Multiple Linear Regression**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.random.rand(100,2)
```

```python
coef=np.array([2,3])
```

```python
y=0+np.dot(x,coef) # x katsayılarla çarpılıp 0 ile toplandı
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection="3d")
ax.scatter(x[:,0],x[:,1],y)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("y")
plt.show()
```

```python
lin_reg=LinearRegression()
```

```python
lin_reg.fit(x,y)
```

```python
x1,x2=np.meshgrid(np.linspace(0,1,10),np.linspace(0,1,10))
```

```python
y_pred=lin_reg.predict(np.array([x1.flatten(),x2.flatten()]).T)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection="3d")
ax.scatter(x[:,0],x[:,1],y)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("y")
ax.plot_surface(x1,x2,y_pred.reshape(x1.shape),alpha=0.2)
plt.title("Multiple Linear Regression")
plt.show()
```

```python
print("Coefficients:",lin_reg.coef_)
print("Intercept:",lin_reg.intercept_)
```

**Bias ve gürültü ekleyerek aynı işlemleri yaparsak**

```python
y1=5+10*np.random.rand(100)+np.dot(x,coef)
```

```python
lin_reg1=LinearRegression()
```

```python
lin_reg1.fit(x,y1)
```

```python
x1,x2=np.meshgrid(np.linspace(0,1,10),np.linspace(0,1,10))
```

```python
y_pred1=lin_reg1.predict(np.array([x1.flatten(),x2.flatten()]).T)
```

```python
fig1=plt.figure()
ax1=fig1.add_subplot(111,projection="3d")
ax1.scatter(x[:,0],x[:,1],y1)
ax1.set_xlabel("x1")
ax1.set_ylabel("x2")
ax1.set_zlabel("y1")
ax1.plot_surface(x1,x2,y_pred1.reshape(x1.shape),alpha=0.2)
plt.title("Multiple Linear Regression")
plt.show()
```

```python
print("Coefficients:",lin_reg1.coef_)
print("Intercept:",lin_reg1.intercept_)
```

**Diabetes veri seti kullanarak çoklu regresyon modeli uygulaması**

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes=load_diabetes()
```

```python
x=diabetes.data
y=diabetes.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
lin_reg2=LinearRegression()
```

```python
lin_reg2.fit(X_train,y_train)
```

```python
y_pred2=lin_reg2.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
rmse=mean_squared_error(y_test,y_pred2,squared=False) # karesi alınmasın
print("RMSE:",rmse)
```
